## 4. Authentication and initializing the API

```python
def build_spotify_client(client_id: str, client_secret: str):
    """
    Build and test a Spotify client using Client Credentials.
    Returns a Spotipy client if successful, otherwise None.
    """
    if not client_id or not client_secret:
        print("Missing CLIENT_ID or CLIENT_SECRET.")
        return None

    try:
        auth_manager = SpotifyClientCredentials(
            client_id=client_id,
            client_secret=client_secret
        )

        sp = spotipy.Spotify(
            auth_manager=auth_manager,
            requests_timeout=10,
            retries=3
        )

        # Lightweight test call
        _ = sp.search(q="Lady Gaga", type="track", limit=1)
        print("Spotify client created successfully.")
        return sp

    except SpotifyException as e:
        print(f"SpotifyException | status={getattr(e, 'http_status', 'N/A')} | message={e}")
    except requests.exceptions.RequestException as e:
        print(f"Network error: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")

    return None


sp = build_spotify_client(CLIENT_ID, CLIENT_SECRET)
```

## 5. Helper functions for safe API calls

```python
def safe_search(sp, q: str, search_type: str = "track", limit: int = 10):
    """
    Safe wrapper for Spotify search.
    Returns the API response dict or an empty dict if something fails.
    """
    if sp is None:
        print("Spotify client is not available.")
        return {}

    try:
        return sp.search(q=q, type=search_type, limit=limit)
    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"SpotifyException during search | status={status} | query='{q}' | message={e}")
        return {}
    except requests.exceptions.RequestException as e:
        print(f"Network error during search: {e}")
        return {}
    except Exception as e:
        print(f"Unexpected error during search: {e}")
        return {}


def extract_track_summary(track: dict) -> dict:
    """
    Extract a clean summary from a Spotify track object.
    """
    if not track:
        return {}

    return {
        "track_name": track.get("name"),
        "track_id": track.get("id"),
        "popularity": track.get("popularity"),
        "uri": track.get("uri"),
        "album_name": track.get("album", {}).get("name"),
        "album_release_date": track.get("album", {}).get("release_date"),
        "artists": ", ".join([a.get("name", "") for a in track.get("artists", [])])
    }
```

### 10. Helper: get artist ID safely

```python
def get_artist_id(artist_name: str):
    """
    Search for an artist and return the first matching artist ID.
    """
    result = safe_search(sp, q=artist_name, search_type="artist", limit=1)
    items = result.get("artists", {}).get("items", [])

    if not items:
        print(f"No artist found for: {artist_name}")
        return None

    return items[0].get("id")
```

### 11. Function: get_top_tracks

```python
def get_top_tracks(artist_name: str, limit: int = 5, market: str = "ES") -> list:
    """
    Return the first `limit` top tracks for a given artist.
    Uses artist ID + artist_top_tracks when available.
    Includes fallback and error handling.
    """
    artist_id = get_artist_id(artist_name)

    if not artist_id:
        return []

    try:
        top_tracks = sp.artist_top_tracks(artist_id, country=market)
        tracks = top_tracks.get("tracks", [])

        track_names = []
        for track in tracks[:limit]:
            name = track.get("name")
            if name:
                track_names.append(name)

        return track_names

    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"SpotifyException | top tracks failed for '{artist_name}' | status={status} | message={e}")

    except requests.exceptions.RequestException as e:
        print(f"Network error while fetching top tracks for '{artist_name}': {e}")

    except Exception as e:
        print(f"Unexpected error while fetching top tracks for '{artist_name}': {e}")

    # Fallback: use search if artist_top_tracks fails
    try:
        result = safe_search(sp, q=f"artist:{artist_name}", search_type="track", limit=10)
        items = result.get("tracks", {}).get("items", [])

        filtered = []
        for t in items:
            artists_in_track = [a.get("name", "").lower() for a in t.get("artists", [])]
            if artist_name.lower() in artists_in_track:
                filtered.append(t.get("name"))

        # Remove duplicates while preserving order
        seen = set()
        cleaned = []
        for t in filtered:
            if t and t.lower() not in seen:
                seen.add(t.lower())
                cleaned.append(t)

        return cleaned[:limit]

    except Exception as e:
        print(f"Fallback search also failed for '{artist_name}': {e}")
        return []
```

### 13. Function: find_related_artists

```python
def find_related_artists(artist_name: str, limit: int = 5) -> list:
    """
    Return the first `limit` related artists for a given artist.
    Uses artist_related_artists when available.
    If it fails, falls back to recommendations seeded by the artist.
    """
    artist_id = get_artist_id(artist_name)

    if not artist_id:
        return []

    # First attempt: artist_related_artists
    try:
        related = sp.artist_related_artists(artist_id)
        items = related.get("artists", [])

        names = []
        for artist in items[:limit]:
            name = artist.get("name")
            if name:
                names.append(name)

        if names:
            return names

    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"artist_related_artists failed for '{artist_name}' | status={status} | message={e}")

    except requests.exceptions.RequestException as e:
        print(f"Network error while fetching related artists for '{artist_name}': {e}")

    except Exception as e:
        print(f"Unexpected error in related artists for '{artist_name}': {e}")

    # Fallback: recommendations
    try:
        recs = sp.recommendations(seed_artists=[artist_id], limit=50)
        tracks = recs.get("tracks", [])

        related_names = []
        seen = set()
        original_lower = artist_name.lower()

        for track in tracks:
            for artist in track.get("artists", []):
                name = artist.get("name", "")
                lower_name = name.lower()

                if lower_name == original_lower:
                    continue
                if lower_name in seen:
                    continue

                seen.add(lower_name)
                related_names.append(name)

                if len(related_names) >= limit:
                    return related_names

        return related_names

    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"recommendations fallback failed for '{artist_name}' | status={status} | message={e}")
    except requests.exceptions.RequestException as e:
        print(f"Network error in recommendations fallback for '{artist_name}': {e}")
    except Exception as e:
        print(f"Unexpected error in recommendations fallback for '{artist_name}': {e}")

    return []
```

### 16. Safe playlist helpers

```python
def safe_featured_playlists(limit: int = 5):
    """
    Try to fetch Spotify featured playlists.
    If the endpoint is unavailable or forbidden, return an empty list.
    """
    try:
        results = sp.featured_playlists(limit=limit)
        items = results.get("playlists", {}).get("items", [])

        cleaned = []
        for pl in items:
            if not pl:
                continue
            cleaned.append({
                "name": pl.get("name"),
                "id": pl.get("id")
            })

        return cleaned

    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"featured_playlists failed | status={status} | message={e}")
        return []

    except requests.exceptions.RequestException as e:
        print(f"Network error while fetching featured playlists: {e}")
        return []

    except Exception as e:
        print(f"Unexpected error while fetching featured playlists: {e}")
        return []
```

```python
def search_playlists_alternative(query: str = "Top Hits", limit: int = 5):
    """
    Fallback method for playlists using search instead of featured_playlists.
    This is useful if featured_playlists is unavailable or restricted.
    """
    result = safe_search(sp, q=query, search_type="playlist", limit=limit)
    items = result.get("playlists", {}).get("items", [])

    cleaned = []
    for pl in items:
        if not pl:
            continue
        cleaned.append({
            "name": pl.get("name"),
            "id": pl.get("id")
        })

    return cleaned
```

### 19. Safe playlist details

```python
def safe_get_playlist_details(playlist_id: str):
    """
    Fetch playlist details safely.
    Returns a dict or None.
    """
    if not playlist_id:
        print("No playlist_id provided.")
        return None

    try:
        playlist = sp.playlist(playlist_id)
        return playlist

    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"Playlist details failed | status={status} | playlist_id={playlist_id} | message={e}")
        return None

    except requests.exceptions.RequestException as e:
        print(f"Network error while fetching playlist details: {e}")
        return None

    except Exception as e:
        print(f"Unexpected error while fetching playlist details: {e}")
        return None
```

### 20. Safe playlist tracks

```python
def safe_get_playlist_tracks(playlist_id: str, limit: int = 10):
    """
    Fetch playlist tracks safely.
    Returns a list of playlist items.
    """
    if not playlist_id:
        print("No playlist_id provided.")
        return []

    try:
        results = sp.playlist_items(playlist_id, limit=limit)
        return results.get("items", [])

    except SpotifyException as e:
        status = getattr(e, "http_status", "N/A")
        print(f"playlist_items failed | status={status} | playlist_id={playlist_id} | message={e}")
        return []

    except requests.exceptions.RequestException as e:
        print(f"Network error while fetching playlist items: {e}")
        return []

    except Exception as e:
        print(f"Unexpected error while fetching playlist items: {e}")
        return []
```
