```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.exceptions import SpotifyException
import requests

# --- AUTH (Client Credentials) ---
CLIENT_ID = "MY_CLIENT_ID"
CLIENT_SECRET = "MY_CLIENT_SECRET"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET
    )
)

# favorite artists
artists = ["Los Fabulosos Cadillacs", "Manu Chao", "Muchachito Bombo Infierno"] 


def safe_top_tracks_by_search(artist_name: str, limit: int = 5) -> list[str]:
    """
    Return the names of the first `limit` top-like tracks for the given artist,
    using the search endpoint (compatible with Client Credentials).
    Includes basic error handling.
    """
    effective_limit = min(max(limit, 1), 10)

    try:
        # Search for tracks with the artist name in the query
        result = sp.search(
            q=f"artist:{artist_name}",
            type="track",
            limit=effective_limit
        )
    except (SpotifyException, requests.exceptions.RequestException) as e:
        print(f"[WARN] Could not fetch tracks for '{artist_name}': {e}")
        return []

    items = result.get("tracks", {}).get("items", [])

    # Filter tracks where the first artist matches the requested artist 
    filtered = [
        t for t in items
        if t.get("artists")
        and t["artists"][0].get("name", "").lower() == artist_name.lower()
    ]

    # Take up to `limit` track names
    top_names = [t.get("name") for t in filtered[:limit]]
    return top_names


# Get the first 5 "top" tracks (via search) for each artist in the list
top_tracks_list = [safe_top_tracks_by_search(artist, limit=5) for artist in artists]

print("Top tracks per artist:")
for artist, tracks in zip(artists, top_tracks_list):
    print(f"\nArtist: {artist}")
    for t in tracks:
        print(f"  - {t}")
```

```python
def find_related_artists(artist_name: str, limit: int = 5) -> list[str]:
    """
    Return the names of the first `limit` artists related to the given artist.
    Uses recommendations with the artist as a seed (Client Credentials compatible).
    """
    # 1. Get artist ID
    search_result = sp.search(q=artist_name, type="artist", limit=1)
    items = search_result.get("artists", {}).get("items", [])
    
    if not items:
        return [f"No artist found for: {artist_name}"]
    
    artist_id = items[0]["id"]
    artist_name_lower = artist_name.lower()

    # 2. Get recommendations using the artist as a seed
    try:
        recs = sp.recommendations(seed_artists=[artist_id], limit=50)
    except Exception as e:
        print(f"[WARN] Could not fetch related artists for '{artist_name}': {e}")
        return []

    tracks = recs.get("tracks", [])
    related = []
    seen = set()

    # 3. Extract unique artist names
    for track in tracks:
        for a in track.get("artists", []):
            name = a.get("name", "")
            name_lower = name.lower()

            if name_lower == artist_name_lower:
                continue
            if name_lower in seen:
                continue

            seen.add(name_lower)
            related.append(name)

            if len(related) >= limit:
                return related

    return related
```
