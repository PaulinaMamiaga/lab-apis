```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.exceptions import SpotifyException
import requests

# --- AUTH (Client Credentials) ---
CLIENT_ID = "MY_CLIENT_ID"
CLIENT_SECRET = "MY_CLIENT_SECRET"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET
    )
)

# favorite artists
artists = ["Coldplay", "Beyoncé", "Radiohead"] 


def safe_top_tracks_by_search(artist_name: str, limit: int = 5) -> list[str]:
    """
    Return the names of the first `limit` top-like tracks for the given artist,
    using the search endpoint (compatible with Client Credentials).
    Includes basic error handling.
    """
    effective_limit = min(max(limit, 1), 10)

    try:
        # Search for tracks with the artist name in the query
        result = sp.search(
            q=f"artist:{artist_name}",
            type="track",
            limit=effective_limit
        )
    except (SpotifyException, requests.exceptions.RequestException) as e:
        print(f"[WARN] Could not fetch tracks for '{artist_name}': {e}")
        return []

    items = result.get("tracks", {}).get("items", [])

    # Filter tracks where the first artist matches the requested artist 
    filtered = [
        t for t in items
        if t.get("artists")
        and t["artists"][0].get("name", "").lower() == artist_name.lower()
    ]

    # Take up to `limit` track names
    top_names = [t.get("name") for t in filtered[:limit]]
    return top_names


# Get the first 5 "top" tracks (via search) for each artist in the list
top_tracks_list = [safe_top_tracks_by_search(artist, limit=5) for artist in artists]

print("Top tracks per artist:")
for artist, tracks in zip(artists, top_tracks_list):
    print(f"\nArtist: {artist}")
    for t in tracks:
        print(f"  - {t}")
```

```python
def get_top_tracks(artist_name):
    # Search for the artist to get their ID
    search_result = sp.search(q=artist_name, limit=1, type="artist")
    
    if len(search_result["artists"]["items"]) == 0:
        return [f"No artist found for: {artist_name}"]
    
    artist_id = search_result["artists"]["items"][0]["id"]
    
    # Fetch top tracks for that artist
    top_tracks = sp.artist_top_tracks(artist_id)
    
    # Extract the first 5 track names
    track_names = [track["name"] for track in top_tracks["tracks"][:5]]
    
    return track_names
```

```python
def get_top_tracks(artist_name):
    search_result = sp.search(q=artist_name, limit=1, type="artist")
    
    if len(search_result["artists"]["items"]) == 0:
        return [f"No artist found for: {artist_name}"]
    
    artist_id = search_result["artists"]["items"][0]["id"]
    
    track_results = sp.search(q=f"artist:{artist_name}", type="track", limit=5)
    
    track_names = [track["name"] for track in track_results["tracks"]["items"]]
    
    return track_names
```

```python
top_tracks_list = [get_top_tracks(artist) for artist in artists]
top_tracks_list
```

```python
def find_related_artists(artist_name):
    # Search for the artist to get their ID
    search_result = sp.search(q=artist_name, limit=1, type="artist")
    
    if len(search_result["artists"]["items"]) == 0:
        return [f"No artist found for: {artist_name}"]
    
    artist_id = search_result["artists"]["items"][0]["id"]
    
    # Fetch related artists
    related = sp.artist_related_artists(artist_id)
    
    # Extract the first 5 related artist names
    related_names = [artist["name"] for artist in related["artists"][:5]]
    
    return related_names
```

```python
def get_top_tracks(artist_name):
    # Search for tracks by this artist
    results = sp.search(q=f"artist:{artist_name}", type="track", limit=5)
    
    # Extract track names
    track_names = [track["name"] for track in results["tracks"]["items"]]
    
    return track_names
```

```python
def find_related_artists(artist_name):
    # Search for artists with similar names
    results = sp.search(q=artist_name, type="artist", limit=10)
    
    # Extract the first 5 artist names (excluding the original)
    names = [a["name"] for a in results["artists"]["items"] if a["name"].lower() != artist_name.lower()]
    
    return names[:5]
```

```python
related_artists_list = [find_related_artists(artist) for artist in artists]
related_artists_list
```

```python
playlist_tracks = []

# Top tracks of favorite artists
for artist in artists:
    playlist_tracks.extend(get_top_tracks(artist))

# Top tracks of related artists
for artist in artists:
    related = find_related_artists(artist)
    for rel_artist in related:
        playlist_tracks.extend(get_top_tracks(rel_artist))
```

```python
print("Generated Playlist:")
for track in playlist_tracks:
    print(track)
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```
